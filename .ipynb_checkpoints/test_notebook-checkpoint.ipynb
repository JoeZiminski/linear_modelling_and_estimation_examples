```python
# TODO: Understanding the Fourier Transform - and it's exciting applications - from scratch.

# Introduction to the basics of the Fourier Transform (e.g. just why it is cool - model signal from sine and cosine)
# Math gets in the way - requires some fundamentals that are not always properly explained. Skipping steps - cannot understand the rest
# Some if its exciting applications - filtering, time shifting, correlations

# Introduction to Sine, Cosine and Functions
# Start by introduce the sine / cosine function - and a minaipulation - how do we get there?
# Introduction to sine, cosine on the unit cirlcle, radians (note pi just a number, show in degree, radian, and number)
# Introduction to the sine, cosine function
# Introduction to a * sin(f * x + ps)

# Signals as an infinite series of sine and cosines

# Complex Numbers, The Exponential Function and the Beauty of the Fourier Transform

# Applying the Fourier Transform (and it's grubby details) (with optional python examples)

# A shift in the time domain is a rotation in the fourier domain
#     S.H. applying a time shift in a single multiplication

# A multlpication in the Fourier domain is a convolution (or correlation!)
# also notes on circularity?

import numpy as np
from matplotlib import pyplot as plt
pi = np.pi 

#
# Define Some functions
#
def set_plot(plt, title):
    plt.xticks([0, (1/2)*pi, pi, (3/2)*pi, 2*pi], ['0', 'π/2', 'π', '3π/2', '2π'])
    plt.legend(loc="upper center",  ncol=4, bbox_to_anchor=(0.5, -0.05))
    plt.title(title)
    
# ---------------------------------------------------------------------------------------
# A rotation in the Fourier Domain is a shift in the time domain.
# ---------------------------------------------------------------------------------------

# Setup the raw data

t = 2*np.pi
x = np.linspace(0, t, 1000)
fs = x.size / t
n = x.size
print(f"Time span of the recording is 2*pi: {t:0.4f} s")
```

```python
# Make 4 Since / Cos waves with different phase amplitude. Recall:
# a * sin(x*f + s) where a is amplitude, f is frequency and s is phase shift

y1 = np.sin(x*5)
y2 = 2 * np.sin(x*10+pi/3)
y3 = 3 * np.cos(x*15+pi/6)
y4 = 1.5 * np.cos(x*20-pi/10)

plt.plot(x, y1, label="y1")
plt.plot(x, y2, label="y2")
plt.plot(x, y3, label="y3")
plt.plot(x, y4, label="y4")
set_plot(plt, "The sine waves that make up our signal")
plt.show()
```

```python
# Add the sine waves together to make a signal
y = y1 + y2 + y3 + y4
plt.plot(x, y, label="y")
#set_plot(plt, "Our signal made up of sine waves")
plt.show()
```

```python
# Now Take the FFT of the signal

# the maximum detectable frequency is half the sampling rate. for each frequency, we have an imaginary
# number and because of complex conjugation we have 2x the frequencies (positive and negative). We can
# ignore the negative frequencies. See below example, the binning algo of fftshift
# is a little tricky to conceptualise but below it can be seen it is essentially
# the fs/n * 0 ... n/2 (with some complication for odd numbers).


Y = np.fft.fftshift(np.fft.fft(y))
freqs = np.fft.fftfreq(x.size, 1/fs)  # this outputs positive then negative frequency bins
freqs = np.fft.fftshift(freqs)        # this does in order negative to positive

example_for_even_numbers = [i * fs/(n) for i in range(0, int((n)/2))]
example_for_even_numbers_2 = np.linspace(-fs/2, fs/2, n + 2)

# Normalisation - to return to actual units, we combine
# a + bi using abs (sqrt(a^2 + b^2)) - if we use Y.real this does not work!
# then divide by n as we have taken the integral, then multiply by 2 because
# we ignore the imaginary half of the spectrum.

norm_Y = (np.abs(Y) / n) * 2
norm_freqs = freqs * 2 * np.pi  # convert to Hz (i.e. the 'recording' lasts 2 * pi seconds)

plt.plot(norm_freqs, norm_Y)
plt.xlim([-25, 25])
plt.xlabel("Frequncy Bin (Hz) (Width: fs/n)")
plt.ylabel("Amplitude")
plt.title("Fourier Transform \n Note the amplitude and frequency matches that of the generates sine waves")
plt.show()
```

### The DC Term

Here the DC term (power at 0 frequency) is zero, however it is important to understand, see [below](https://dsp.stackexchange.com/questions/12972/discrete-fourier-transform-what-is-the-dc-term-really):

"The DC term is the 0 Hz term and is equivalent to the average of all the samples in the window
hence it's always purely real for a real signal). The terminology does indeed come from AC/DC electricity
-all the non-zero bins correspond to non-zero frequencies, i.e. "AC components" in an electrical context,
whereas the zero bin corresponds to a fixed value, the mean of the signal, or "DC component" in electrical terms."

We can see that when f = 0 in the fourier transform itnergral then
this reduces to integral over x(t) which is the mean after normalisation (see above, divide by N)


$$ \int x(t) e^{-i\pi f\frac{n}{N}} dx $$

reduces to $\int x(t) dx $ when f = 0.

In our example it is zero because the average of sine  and cosines are zero. In nosiey signals this is not the case 
and DC offset can be large.

As far as practical applications go, the DC or 0 Hz term is not particularly useful.
In many cases it will be close to zero, as most signal processing applications will
tend to filter out any DC component at the analogue level.
In cases where you might be interested it can be calculated directly as an average in the usual way, without resorting to a DFT/FFT.

### Extracting the Phase

We have wound our signal around the complex plane (see [here](https://www.youtube.com/watch?v=spUNpyF58BY&vl=en) and [here](https://github.com/morphx666/VisualFT) for fantastic visualisations).

A number of the complex plane can be written as $a + bi$ where a is the real axis (i.e. cartesian x) and i the imaginary axis (i.e. cartesoian y). Alternatively, it can be written in modulus argument form where $mod = \sqrt{a^2 + b^2}$ and $arg = \arctan\frac{b}{a}$. Here mod is the euclidean distance between the complex number and the origin, while arg is the angle rotation from the real axis. 

For each frequency, we get a complex number representing a point on the complex plane. When we take the aboslute value when normalising (absolute value, np.abs()) we are taking the modulus. This represents the power in the signal (i.e. the amplitude of a sine wave at that frequency). Alternatively we can take the argument, i.e. the phase of the sine wave.

Using np.angle to do this, we can see the phase of the sine wave at each frequency detected.

```python
# Plotting the phase term
plt.plot(norm_freqs, np.angle(Y), label="Frequency (Hz)")
plt.plot(norm_freqs, norm_Y, label="Phase (Angle in Radians)")
plt.xlim([0, 200])
plt.xlabel("Frequency Bin (Hz)")
plt.legend()
plt.title("The Signal Phase Overlaid on Frequency")
plt.show()

plt.plot(norm_freqs, np.angle(Y), label="Frequency (Hz)")
plt.plot(norm_freqs, norm_Y, label="Phase (Angle in Radians)")
plt.xlim([0, 25])
plt.xlabel("Frequency Bin (Hz)")
plt.legend()
plt.title("Zoomed: The Signal Phase Overlaid on Frequency")
plt.show()
```

Now, the units here have clear meaning. Lets compare this plot to the phase of the sine waves we origionally generated 

So the phase is measured from cos(0) not sin(0). We can see from the plot (estimated):

$y1 = \sin(5x)$     
magnitude = 1, frequency = 5, phase offset (from $cos{0}$) = 0 + 90 degree = pi / 2 = 1.5707

$ y2 = 2\sin(10x + \frac{\pi}{3})$   
magnitude = 2, frequency = 10,  phase offset (from $cos{0}$)  = pi/3 + pi/2 = 2.617

$ y2 = 2\cos(15x + \frac{\pi}{6})$   
magnitude = 3, frequency = 15,  phase offset (from $cos{0}$)  = pi / 6 = 0.523

$ y2 = 2\cos(20x + \frac{\pi}{10})$   
magnitude = 1.5, frequency = 20,  phase offset (from $cos{0}$)  = pi / 10 = 0.314

#### NOTE:
The complex representation of frequency is such that the real part corresponds to a cosine component and the imaginary part to a sine component.
So a complex phase of 0 corresponds to a cosine wave, not a sine wave.
This is why the computed phases are off by about 90 degrees from what you expect,
according to the trig identity sin(x) = cos(x − π/2). The lower-frequency wave should have a
phase of −90 degrees and the higher-frequency wave should have a phase of 30 − 90 = −60 degrees.


## Applying a rotation in the frequency domain

Now, apply a rotation by multiplying the fourier transform by exponential with shift linearly
proportional to the frequency. The change must be linearly proprotional to frequency rather
than the same for all frequencies, because e.g. a half-period of a fast sine wave is shorter
than a slow sine wave.


```python
Y_rot = Y * np.exp(-1j * 2 * np.pi * freqs * np.pi/2)
plt.plot(norm_freqs, (np.abs(Y_rot)/ n ) * 2, label="Frequency (Hz)")
plt.plot(norm_freqs, np.angle(Y_rot), label="Phase (Angle in Radians)")
plt.xlim([0, 200])
plt.title("After Rotation - The Signal Phase Overlaid on Frequency")
plt.show()
```

```python
# Finally, we see the 
# See, the shift has been applied! amazing!! beautiful!
y_rot = np.real(np.fft.ifft(np.fft.ifftshift(Y_rot)))
plt.plot(x, y, label="original y")
plt.plot(x, y_rot, label="shifted y")
#set_plot(plt, "y before and after shifting")
plt.show()
```
